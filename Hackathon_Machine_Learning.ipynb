# Logistic Regression 

```python
X=df.iloc[:,:-1].values
y=df['Result']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,mean_squared_error
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
model1=LogisticRegression()
model1.fit(X_train,y_train)
```

```python
model1= tree.DecisionTreeClassifier(random_state=2)
model1.fit(X_train,y_train)
```

```python
prediction=model1.predict(X_test)
print("Accuacy Score for Test--->  {:.2f} " .format(model1.score(X_test,y_test)))
```

# Confusion Matrix

```python
matrix=confusion_matrix(y_test,prediction)
print(matrix)
```

#  Mean Square Error

```python
import math
print(math.sqrt(mean_squared_error(y_test,prediction)))
```

# Performance on train set

```python
y_train_predict=model1.predict(X_train)
print(classification_report(y_train,y_train_predict))
```

```python
confusion_matrix(y_train,y_train_predict)
```

```python
sns.heatmap(confusion_matrix(y_train,y_train_predict),annot=True,cmap='PuBuGn',linewidths=.5,square=True)
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.title("Accuracy Score:"+str(round(model1.score(X_train,y_train)*100,2))+'%')
plt.show()
```

# Performance on test set

```python
y_test_predict=model1.predict(X_test)
print(classification_report(y_test,y_test_predict))
```

```python
confusion_matrix(y_test,y_test_predict)
```

```python
sns.heatmap(confusion_matrix(y_test,y_test_predict),annot=True,cmap='PuBuGn',linewidths=.5,square=True)
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.title("Accuracy Score:"+str(round(model1.score(X_test,y_test)*100,2))+'%')
plt.show()
```

# Comparing both the models 

```python
def compare_train_test(models):
    y_train_predict=model1.predict(X_train)
    y_test_predict=model1.predict(X_test)
    cr_train=classification_report(y_train,y_train_predict)
    cr_test=classification_report(y_test,y_test_predict)
    print(models.score(X_train,y_train).round(2),cr_train,'\n',model1.score(X_test,y_test).round(2),cr_test,'\n')
compare_train_test(model1)
```

# ROC Curve

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(y_test, model1.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, model1.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'g--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend(loc="lower right")
plt.show()
```
